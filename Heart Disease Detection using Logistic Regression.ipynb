Data Collection,Analysis and Processing

```python
#loading the heart disease dataset
heart_data = pd.read_csv('/content/heart.csv')
```

Splitting the Feaures and target

```python
X = heart_data.drop(columns='target', axis=1)
Y= heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, stratify=Y, test_size= 0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with training data

model.fit(X_train,Y_train)
```

MOdel Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction= model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data:---',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction= model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data:---',test_data_accuracy)
```

Building a Predictive System

```python
input_data = (41,0,1,130,204,0,0,172,0,1.4,2,0,2)

#we need to change input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data) 

#reshaping the array to tell the machine learning model that i want to predict the target for only one variable
# as we predicting for onlyy one instance

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)
```
