Train Test Split

```python
# if we dont use the stratify then all the data for nondiabetic or diabetic might go to the train data or test data i.e., 
#to get the same proportions of diabetic and nondiabetic in both train and test data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2 )
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the Model

```python
classifier= svm.SVC(kernel='linear')
```

```python
#training the Support Vector Machine classifier
classifier.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuraacy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data:-- ',training_data_accuracy)
```

```python
#accuraacy score on the test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the training data:-- ',test_data_accuracy)
```

Making a Predictive System

```python
input_data=(4,110,92,0,0,37.6,0.191,30)

#changing the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshaping the array as we are predicting for one array 
# we reshaped the array to let our model know that we are using a single instance not multiple values
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

#since our classifier returns a list and not an integer therfore we used prediction[0]
if (prediction[0] == 0):
  print('The Person is not Diabetic')
else:
  print('The Person is Diabetic')
```
