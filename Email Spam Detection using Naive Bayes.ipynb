```python
#import the libraries
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
import string
```

```python
# read the csv

data= pd.read_csv('/content/drive/MyDrive/Kaggle Dataset/spam.csv', encoding="latin-1")
```

```python
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],axis=1, inplace=True)
```

```python
data.head()
```

```python
data['class']=data['class'].map({'ham':0, 'spam':1})
```

```python
data.head()
```

```python
#Data Preprocessing 
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv= CountVectorizer()
```

```python
x= data['message']
y= data['class']
```

```python
x.shape
```

```python
y.shape
```

```python
x=cv.fit_transform(x)
```

```python
x
```

```python
# #This kind of sparse matrix it will generate(ountVectorizer)
# 1.The Cat
# 2.The Dog
# 3.The Bird

#   The Cat Dog Bird
# 1. 1  1   0     0
# 2. 1  0   1     0
# 3. 1  0   0     1
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model= MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
result = model.score(x_test,y_test)
```

```python
result= result*100
```

```python
result
```

```python
import pickle
```

```python
pickle.dump(model, open("spam.pkl","wb"))
```
